```python
import json 
import gmaps
import os
import gmaps.datasets
import googlemaps
import pandas as pd
```

```python
bayArea = pd.read_csv("bayarea_zipcodes.csv")
bayArea_zipcodes = bayArea["ZIP"]
```

```python
gmaps = googlemaps.Client(key=#INSERT API KEY#)


company_zipCode = 94039 # Mountain View #

durations = []
for x in range(len(bayArea_zipcodes)):
    metrics = gmaps.distance_matrix(str(company_zipCode), str(bayArea_zipcodes[x]))['rows'][0]
    commuteTime = metrics['elements'][0]["duration"]['value'] / 60
    durations.append(commuteTime)
```

```python
bayArea["Avg Commute Times"] = durations
durations
```

```python
zipsToLats = pd.read_csv("zipToLats.csv", sep=";")
zipsToLats
```

```python
Lats = []
Longs = []
for x in range(len(bayArea_zipcodes)):
    row = zipsToLats.loc[(zipsToLats["Zip"] == bayArea_zipcodes[x])]
    Lats.append(float(row["Latitude"]))
    Longs.append(float(row["Longitude"]))
    

Lats
```

```python
bayArea["Latitude"] = Lats
bayArea["Longitude"] = Longs
```
